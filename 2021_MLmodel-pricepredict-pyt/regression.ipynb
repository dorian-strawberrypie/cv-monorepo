
# Ordinary least squares regression using Stochastic gradient descend

## Predicting housing pricing from Boston dataset
1a. Use **csv_interface** to load data from **boston.csv**

```python
path = 'datasets/boston.csv'

X_train, y_train, X_test, y_test = read_from_csv(path, 'medv', header_row=0, indexer_column=0)
```

1b. plot parameter values

```python
# dsitribution of parameter Xi on y
i = 0

plt.title("Xi on y distribution")
plt.plot(X_test[:,i], y_test, 'o')
```

2. Start **OLS model** and **SGD**

```python
max_iter = 1000
tolerance = 1e-5
learning_rate = .04
weight_coefficient = -0.5

model = OLS(max_iter, tolerance, learning_rate)
model.starting_weight = weight_coefficient
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
squared_error = score(y_test, y_pred)

print('Squared error: ', squared_error)

plt.figure()
plt.title("Predictions vs actual values")
plt.plot(X_test[:,i], y_pred, 'o')
plt.plot(X_test[:,i], y_test, 'o')

```
